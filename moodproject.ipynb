```python
from transformers import pipeline
```

```python
model_checkpoint = "Tatyana/rubert-base-cased-sentiment-new"
classifier = pipeline("text-classification", model=model_checkpoint)
```

```python
classifier("Аналитикой не занимаюсь  - просто  посмотрите Графики перемещения и встреч Лидеров стран СНГ и ЮВА.Сложите свой Пазл.Это полный дурдом,они не понимают что делают.")
```

```python
classifier("Молодая пара Дона и Ненси совершает вынужденную остановку у обочины небольшого городка. Неприветливый Шериф сообщает, что до утра их автомобиль отремонтировать никто не сможет. Устроившись в захудалом мотеле они и не подозревают, что эта ночь может стать последней в их жизни... ")
```
